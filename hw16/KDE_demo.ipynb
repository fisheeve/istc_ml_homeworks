```python
import numpy as np
import matplotlib.pyplot as plt
from KDE import KDE, gaussian, epaneshnikov, uniform
```

```python
data1 = np.array([np.random.normal(0,1,size=500)])
data2 = np.array([np.random.normal(5,1,size=1000)])
data = np.hstack((data1,data2)).flatten()
```

```python
plt.hist(data, bins=100, density=True);
```

## uniform kernel

```python
model = KDE(0.5)
```

```python
model.train(data)
```

```python
y = np.arange(-2, 8, 0.01)
prediction = np.array([model.predict(x) for x in y])
```

```python
plt.plot(y, prediction)
plt.hist(data, bins=100, density=True);
```

## gaussian kernel

```python
model_g = KDE(0.5, kernel=gaussian)
model_g.train(data)
    
```

```python
y = np.arange(-2, 8, 0.02)
prediction_g = np.array([model_g.predict(x) for x in y])
```

```python
plt.plot(y, prediction_g)
plt.hist(data, bins=100, density=True);
```

# epaneshnikov kernel

```python
model_e = KDE(1, kernel=epaneshnikov)
model_e.train(data)
```

```python
y = np.arange(-2, 8, 0.02)
prediction_e = np.array([model_e.predict(x) for x in y])
plt.plot(y, prediction_e)
plt.hist(data, bins=100, density=True);
```
